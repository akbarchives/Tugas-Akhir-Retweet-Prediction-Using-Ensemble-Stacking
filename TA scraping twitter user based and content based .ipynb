```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import re
# from google_trans_new import google_translator
# translator = google_translator()
# translator = google_translator(url_suffix="hk",timeout=5,proxies={'http':'45.140.13.119'})
from textblob import TextBlob
```

# 1. scraping using sns

```python
query = "(kominfo)-filter:replies lang:id until:2022-07-17 since:2022-4-17"
tweets = []
limit = 20000

%time
try:
    print("start scraping")
    for tweet in sntwitter.TwitterSearchScraper(query=query).get_items():
        if len(tweets) == limit:
            break
        else:
            tweets.append([tweet.date, tweet.id, tweet.user.username, len(tweet.user.username), tweet.content, len(tweet.content), tweet.mentionedUsers, tweet.replyCount, tweet.retweetCount, tweet.quoteCount, tweet.likeCount,  tweet.outlinks, tweet.media, tweet.user.followersCount, tweet.user.friendsCount, tweet.user.statusesCount, tweet.user.favouritesCount, tweet.user.verified, tweet.user.created])
    df = pd.DataFrame(tweets, columns=['datetime', 'tweet id', 'username', 'username_len', 'content', 'len_of_text', 'mentioned user', 'number_of_reply', 'number_of_retweet', 'number_of_quote_retweet','number_of_like', 'outlinks', 'media', 'no_follower', 'no_following', 'total_tweet','no_of_favorite','verified','created'])
except Exception as e:
    print(e)


print("Finished")
print("--------")
```

```python
# data cleaning

tweet_bersih = []
tweet_content = df['content']
for data in range(0,len(tweet_content)):
    tweet = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet_content.loc[data]).split())
    tweet_bersih.append(tweet)
    
tweet_bersih = pd.DataFrame({'cleaned_tweet':tweet_bersih})

tweet_bersih
```

## 2.2 labeling for age_account

```python
from datetime import date
from datetime import datetime
now = date.today()

def numOfDays(date1, date2):
    return (date2-date1).days
```

```python
df_created = pd.to_datetime(df['created'], format='%Y-%m-%d').dt.date
df_created
df_age = []
for data in range(0,len(df_created)):
    x = (numOfDays((df_created.loc[data]),now))
    df_age.append(x)

df['age_of_account'] = df_age
df
```

## 2.3 labeling for avg_tweet_per_day

```python
df_no_tweet = df[['total_tweet','age_of_account']]
df2 =[]

for index, row in df_no_tweet.iterrows():
    calc = row['total_tweet'] / row['age_of_account']
    df2.append(calc)

df['avg_tweet_per_day'] = df2

df.head()
```

## 2.4 labeling for avg_fav_per_day

```python
df_no_fav = df[['no_of_favorite','age_of_account']]
df2 =[]

for index, row in df_no_fav.iterrows():
    calc = row['no_of_favorite'] / row['age_of_account']
    df2.append(calc)

df['avg_fav_per_day'] = df2

df
```

## 2.5 labeling for contain_picture & contain_video

```python
df_media = df['media']
df_photo = []
df_video = []
for data in range(0,len(df_media)):
    contain = df_media.loc[data]
    if contain == None:
        df_photo.append(0)
        df_video.append(0)
    else:
        toString = str(contain)
        if re.search(r'(?=.*Video\b)', toString):
            df_video.append(1)
            df_photo.append(0)
        elif re.search(r'(?=.*Photo\b)', toString):
            df_photo.append(1)
            df_video.append(0)
        else:
            df_photo.append(0)
            df_video.append(0)

df['contain_picture'] = pd.DataFrame(df_photo)
df['contain_video'] = pd.DataFrame(df_video)

df
```

## 2.6 labeling for contain_link

```python
df_outlinks = df['outlinks']
contain_link = []
contain = []
max = 3
for data in range(0,len(df_outlinks)):
    contain = df_outlinks.loc[data]
#     print(contain)
    if len(str(contain)) > max:
        contain_link.append(1)
    else:
        contain_link.append(0)

df['contain_link'] = pd.DataFrame(contain_link)
df
```

## 2.7 labeling contain hashtag

```python
df_content = df['content']
df_hashtag = []
for data in range(0,len(df_content)):
    contain = df_content.loc[data]
    if re.search(r'(?=.*#\b)', contain):
        df_hashtag.append(1)
    else:
        df_hashtag.append(0)

df['contain_hashtag'] = pd.DataFrame(df_hashtag)

# df
```

## 2.8 labeling contain Upper

```python
df_content = df['content']
df_upper = []
bol = False

for data in range(0,len(df_content)):
    content = df_content.loc[data]
    bol = False
    for ele in content:
        if ele.isupper():
            bol = True
    if bol == True:
        df_upper.append(1)
    else:
        df_upper.append(0)

df['contain_upper'] = pd.DataFrame(df_upper)

df
```

## 2.9 labeling contain user mentioned

```python
df_content = df['content']
df_mention = []
for data in range(0,len(df_content)):
    contain = df_content.loc[data]
    if re.search(r'(?=.*@\b)', contain):
        df_mention.append(1)
    else:
        df_mention.append(0)

df['con_user_mentioned'] = pd.DataFrame(df_mention)
# df
```

## 2.10 labeling opt_length

```python
df_len = df['len_of_text']
df_opt_length = []
for data in range(0,len(df_len)):
    if df_len.loc[data] > 70 and df_len.loc[data] <= 100:
        df_opt_length.append(1)
    else:
        df_opt_length.append(0)
        
df['opt_length'] = pd.DataFrame(df_opt_length)
# df
```

## 2.11 labeling contain Retweet

```python
df_RT = df['number_of_retweet']
df_retweeted = []
for data in range(0,len(df_RT)):
#     print(df_len.loc[data])
    if df_RT.loc[data] > 0:
        df_retweeted.append(1)
    else:
        df_retweeted.append(0)
        
df['retweeted'] = pd.DataFrame(df_retweeted)
df
```

## 2.12 labeling contain verified

```python
df_content = df['verified']
df_verified = []
for data in range(0,len(df_content)):
    if df_content.loc[data] == True:
        df_verified.append(1)
    else:
        df_verified.append(0)
        
df['verified'] = pd.DataFrame(df_verified)
df
```
